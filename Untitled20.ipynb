```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_wine
wine = load_wine()
%matplotlib inline
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
df = pd.DataFrame(wine.data,columns=wine.feature_names)
df.head()
```

```python
df['target'] = wine.target
df.head()
```

```python
df['Class'] = df.target.apply(lambda x: wine.target_names[x])
df.head()
```

```python
df.tail()
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
plt.xlabel('Alcohol')
plt.ylabel('Malic_acid')
plt.scatter(df0['alcohol'], df0['malic_acid'],color="green",marker='+')
plt.scatter(df1['alcohol'], df1['malic_acid'],color="blue",marker='.')
```

```python
plt.xlabel('Ash')
plt.ylabel('Total Phenols')
plt.scatter(df0['ash'], df0['total_phenols'],color="green",marker='+')
plt.scatter(df1['ash'], df1['total_phenols'],color="blue",marker='.')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','Class'], axis='columns')
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print(len(X_train), len(X_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
error1= []
error2= []
for k in range(1,15):
    knn= KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred1= knn.predict(X_train)
    error1.append(np.mean(y_train!= y_pred1))
    y_pred2= knn.predict(X_test)
    error2.append(np.mean(y_test!= y_pred2))
plt.plot(range(1,15),error1,label="train")
plt.plot(range(1,15),error2,label="test")
plt.xlabel('k Value')
plt.ylabel('Error')
plt.legend()
```
